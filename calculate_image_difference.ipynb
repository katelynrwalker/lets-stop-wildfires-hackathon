```python
current_image_number = 0
```

```python
for i in range(0, len(df)):
    if i % 100 == 0:
        print(i)
    if i >= current_image_number:
        temp1 = image.imread(f'img/20170625-BBM-bm-n-mobo/{df.image1[i]}')
        temp2 = image.imread(f'img/20170625-BBM-bm-n-mobo/{df.image2[i]}')
        temp = image.fromarray(temp2-temp1)
        temp.save(f'diff/{df.new_name[i]}', format='JPEG')
```

```python
for i in range(0, len(df)):
    if i % 100 == 0:
        print(i)
    if i >= current_image_number:
        temp1 = Image.open(f'img/20170625-BBM-bm-n-mobo/{df.image1[i]}')
        temp2 = Image.open(f'img/20170625-BBM-bm-n-mobo/{df.image2[i]}')
        temp = asarray(temp2)-asarray(temp1)
        temp = Image.fromarray(temp)
        temp.thumbnail((128,128))
        temp.save(f'diff/20170625-BBM-bm-n-mobo/{df.new_name[i]}', format='JPEG')
```

```python
def calculate_image_diff128():
    for i in range(0, len(df)):
        if i % 100 == 0:
            print(i)
        if i >= current_image_number:
            try:
                temp1 = Image.open(f'img/{df.image1[i]}')
                temp2 = Image.open(f'img/{df.image2[i]}')
                temp = asarray(temp2)-asarray(temp1)
                temp = Image.fromarray(temp)
                temp.thumbnail((128,128))
                folder = 'smoke' if df.label[i] == 1 else 'no_smoke'
                temp.save(f'diff/{folder}/{df.new_name[i]}', format='JPEG')
            except:
                pass
```

```python
calculate_image_diff128()
```
